```python
import numpy as np
import pandas as pd

import mlframework
```

```python
# import data
data = pd.read_parquet('../../aai_banca_ls_v3/data/training/hub')
data['AGE'] = data['AGE'].astype(float)
data[['GENDER', 'CITY']] = data[['GENDER', 'CITY']].astype('category')
data.info()
data.head()
```

```python
feat_cols = ['GENDER', 'CITY', 'AGE', 'MONTH_ON_BANK', 'CASA_BAL_MEAN_BAL_INCREASE_GREATER_1E6_6M']
```

```python
ml = mlframework.MLFramework()
# split data
ml.train_valid_test_split(
    data=data,  # pandas data frame having CUSID and REF_TIME
    label_col='LABEL',  # label column
    feat_cols=feat_cols,  # feature columns
    valid_size=0,
    test_size=0.2
)
# tune model
ml.tune(n_trials=5, cv=True, cv_method='random', nfold=3)
```
